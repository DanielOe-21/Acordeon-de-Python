### Funciones
Hacen operaciones sobre los objetos, ejemplo: ```len()``` y ```type()```. Se identifican porque llevan los paréntesis y dentro de ellos se ponen las entradas.

Una función es un determinado código reunido bajo un nombre que realiza una serie de operaciones. Recibe unos datos de entrada (parámetros de entrada) y devuelve unos datos de salida (salida):

"Entrada" ---> | FUNCION | ---> "Salida" 

Ejemplo: ```len('palabras')```

#### Estructura de una función:

~~~~
def nombre(parámetros): 
Cuerpo de la función 
return objeto
~~~~

**def** indica que estamos creando una función

**nombre** sera el nombre que usaremos para llamar la función más adelante nuestro código

**parámetros** entradas necesarias para que nuestro código realice ciertas operaciones, es opcional

**Cuerpo** Aquí se especifica que las acciones a realizar con los parámetros

**return** Devuelve el objeto que se indica para poder ser usado en nuestro código, es opcional

**Ejemplo** 

* Con parámetros y sin valores de retorno

~~~~

def saludar(nombre):
	print("hola ", nombre, ", integrante de la comunidad EDteam")

#Ejecutar la función
saludar("pedro")
saludar("juan")
saludar("maria")
saludar("claudia")

~~~~

* Con parámetros y con valores de retorno

~~~~

def saludar(nombre):
	return "hola " + nombre + ", integrante de la comunidad EDteam"

# Ejecutar la función
resultado = saludar("pedro")
print(resultado)
resultado = saludar("maria")
print(resultado)

~~~~

#### Sets

* Colección desordenada de elementos únicos.
* Es iterable y mutable, se pueden añadir elementos y modificar.
* Se eliminan elementos duplicados.
* Al ser desordenados, no se pueden extraer elementos dentro de ellos.
* Los elementos se encuentran contenidos dentro de llaves {} y son separados por comas.

Definición de un conjunto(set), elimina los duplicados:

~~~

print(my_set)
{2, 'Dats', 'python', 'Science', 'Datas', 'Data'}
~~~~ 

##### Operaciones con Sets
~~~
#Definimos 2 sets

{2, 4}

#Encontrar los valores que tiene A diferentes que B

{0, 6, 8}

#Encontrar los valores que tiene B diferentes que A

#También con resta se puede
B-A
{1, 3, 5}

#Obtiene un elemento de forma aleatoria del set y lo elimina
A.pop()
{2, 4, 6, 8}

#Si quisieramos guardar ese valor entonces haríamos lo siguiente:
valor = A.pop()
print(A)
print(valor)

#Se puede añadir un valor

{0, 2, 4, 8, 10}

#Se puede remover un valor
A.remove(8)
{0, 2, 4, 10}
~~~

###  Condicional if

La instrucción "if" equivale al "si" condicional. Es usado para ejecutar determinadas intrucciones en bloque si se cumple una determinada condición representada por una expresión booleana.

1. if simple
2. varios if simple anidados
3. if-else
4. Varios if-else anidados
5. if - elif - else
6. Parte de expresiones condicionales

Ejemplo: Si *x* vale 3, entonces *y* vale 10.

~~~

x = 3
if x = 3:
    print('"y" es igual a 10')
    
~~~

#### El formato del if simple es:

~~~

#### El if - else

Los *if simples* realizan una serie de acciones si la condición es verdadera, pero no hacen nada si es falsa. La instrucción combinadad *if - else* añade la posibilidad de ejecutar un bloque de instrucciones cuando la condición es falsa.

~~~
if -CONDICION-:
(TAB) -ACCIÓN-
else:
(TAB) -ACCIÓN-


num = 10
if num > 0:
    print("El número es positivo y distinto de cero")
else:
    print("El número es negativo o cero")
~~~

### Ciclos o bucles

Los ciclos permiten ejecutar de forma reiterada un bloque de código. En algunas ocasiones sabremos el número exacto de repeticiones, a veces la condición que hace que salgamos del bucle.

* *for* : El bucle es controlado mediante una variable contador.
* *while* : El bucle lo controla una condición lógica.

#### Ciclo for
Con esta instrucción creamos un bucle controlado por una variable que recorre una serie de valores.

~~~
for *variable_contador* in *secuencia_de_valores*:
    *bloque de instrucciones*
~~~

La **variable_contador** va pasando por una **secuencia_de_valores** que debemos indicar de algún modo (por ejemplo con la función ``range()``.

Ejemplo:

~~~~
suma = 0  #Inicializar la variable

for i in range(1,9):   #range simula una lista de números
    suma = suma + i # suma+= i
print(suma)

#Otro ejemplo:

nombres = ['Juan','Alberto','Pedro','Rodrigo']

for i in nombres:
    print(i)
    
#De esta forma estamos recorriendo una lista de nombres
~~~~

#### Ciclo while

Con esta instrucción entramos y nos mantenemos en un bucle si se cumple una determinada condición lógica.

````

while *condicion_logica_bucle*:
    *bloque de instrucciones*

````

La **condición_lógica_bucle** es una expresión booleana. Si es True, ejecutará el *bloque de instrucciones*, después volverá a comprobar la condición lógica hasta que ésta sea falsa y salga del *while*.

Ejemplos:
~~~~
for i in range(0,7)
print(i)

###########

i=0
while i<7:
    print(i)
    i=i+1
~~~~

En el ciclo *while* existe el riesgo de entrar en un bucle infinito al cumplirse siempre la condición de mantenimiento en él.
Si en el ejemplo nos olvidamos de actualizar la variable i, la condición del *while* siempre se cumpliría y "se colgaría".

Podemos usar sentencias ```break, continue y pass``` en nuestros bucles para agregar funcionalidad adicional para varios casos. Las tres declaraciones se definen por:
```
break: rompe el bucle envolvente más cercano actual.
continue: va a la parte superior del bucle circundante más cercano.
pass: No hace nada en absoluto.
```
